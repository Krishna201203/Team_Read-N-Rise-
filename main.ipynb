```python

import mysql.connector as a
#import PIL.Image

rnr=a.connect(host='localhost',user='root',passwd='221188',database='LIBRARY_MANAGEMENT_SYSTEM')
def addbook():
    bn=input("Enter Book Name:")
    ba=input("Enter Author's Name:")
    c=int(input("Enter Book Code:"))
    t=int(input("Total Books:"))
    s=input("Enter Subject:")
    data=(bn,ba,c,t,s)
    sql='insert into books_ values(%s,%s,%s,%s,%s);'
    c=rnr.cursor()
    c.execute(sql,data)
    rnr.commit()
    print("Book Added Successfully..........")
    wait=input('Press enter to continue.....')
    main()
    
    
def issueb():
    n=input("Enter Student Name:")
    r=int(input("Enter Reg NO.:"))
    co=int(input("Enter Book Code:"))
    d=input("Enter Date:")
    a="insert into issue_ values(%s,%s,%s,%s);"
    data=(n,r,co,d)
    c=rnr.cursor()
    c.execute(a,data)
    rnr.commit()
    print("Book Issuued Successfully to: ",n)
    wait=input('\nPress enter to continue.....')
    bookup(co,-1)
    main()
    
    
def returnb():
    n=input("Enter Student Name:")
    r=int(input("Enter Reg NO.:"))
    co=int(input("Enter Book Code:"))
    d=input("Enter Date:")
    a="insert into return___ values(%s,%s,%s,%s);"
    data=(n,r,co,d)
    c=rnr.cursor()
    c.execute(a,data)
    rnr.commit()
    print("Book returned by: ",n)
    wait=input('\nPress enter to continue.....')
    bookup(co,1)
    main()
    
    
    
def bookup(co,u):
    a="Select total from books_ where bcode=%s;"
    data=(co,)
    c=rnr.cursor()
    c.execute(a,data)
    myresult=c.fetchone()
    t=myresult[0]+u
    sql="update books_ set total=%s where bcode=%s;"
    d=(t,co)
    c.execute(sql,d)
    rnr.commit()
    c.close()
    wait=input('Press enter to continue.....')
    main()
    
    
    
def dbook():
    ac=int(input("Enter Book Code:"))
    a="delete from books_ where bcode=%s;"
    data=(ac,)
    c=rnr.cursor()
    c.execute(a,data)
    rnr.commit()
    print("Book deleted Successfully")
    wait=input('Press enter to continue.......')
    main()
    
    
    
def dispbook():
    a="select * from books_;"
    c=rnr.cursor()
    c.execute(a)
    myresult=c.fetchall()
    for i in myresult:
        print("Book Name:",i[0])
        print("Author:",i[1])
        print("Book Code:",i[2])
        print("Total:",i[3])
        print("Subject:",i[4])
        print('\n')
        
    wait=input('Press enter to continue.......')
    main()
    
    
    
def report_issued_books():
    a="select * from issue_;"
    c=rnr.cursor()
    c.execute(a)
    myresult=c.fetchall()
    for i in myresult:
        print(myresult)
        
        
    wait=input('Press enter to continue.......')
    main()
    
    
    
    
def report_return_books():
    a="select * from return___;"
    c=rnr.cursor()
    c.execute(a)
    myresult=c.fetchall()
    for i in myresult:
        print(myresult)
        
        
    wait=input('Press enter to continue.......')
    main()
    
    
    
def main():
    print("          WELCOME TO LIBRARY MANAGEMENT SYSTEM ")
    #img = PIL.Image.open("C:/Users/krish/Downloads/LMS.png")
    #img.show()
    #print(img)
    
    
    print("""
             LIBRARY MANAGEMENT SYSTEM
           -----------------------------
           
           1. ADD BOOK
           2. ISSUE OF BOOK
           3. RETURN OF BOOK
           4. DELETE BOOK
           5. DISPLAY BOOKS
           6. REPORT MENU
           7. EXIT PROGRAM
            """)
    choice=input("Enter Task No:.......")
    print('\n')
    if(choice=='1'):
        addbook()
    elif(choice=='2'):
        issueb()
    elif(choice=='3'):
        returnb()
    elif(choice=='4'):
        dbook()
    elif(choice=='5'):
        dispbook()
    elif(choice=='6'):
        print(''' REPORT MENU
                ---------------
                
               1. ISSUED BOOKS
               2. RETURNED BOOKS
               3. GO BACK TO MAIN MENU
               ''')
        choice=input("Enter Task No:.......")
        print('\n')
        if choice=='1':
            report_issued_books()
        elif choice=='2':
            report_return_books()
        elif choice=='3':
            main()
        else:
            print("Please try again..........")
            main()
    elif(choice=='7'):
        print('Thank you and have a great day ahead...!!!')
    else:
        print("Please try again.......")
        

main()
    

    
```
